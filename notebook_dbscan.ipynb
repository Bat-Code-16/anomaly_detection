```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import DBSCAN
import pandas as pd

# Assuming dd is your DataFrame with columns 'OC', 'PC', 'OT'
scaler = MinMaxScaler()

# Fit the scaler to the data
scaler.fit(dd[["OC", "PC", "OT"]])

# Transform the data using the fitted scaler
scaled_data = scaler.transform(dd[["OC", "PC", "OT"]])

# Apply DBSCAN on the scaled data
dbscan = DBSCAN(eps=0.1, min_samples=3)
dd['cluster'] = dbscan.fit_predict(scaled_data)

# Now dd contains a new column 'cluster' with the cluster labels assigned by DBSCAN
print(dd)


```

```python
fig = px.scatter(
    dd,
    x='OC',
    y='PC',
    color=dd['cluster'].astype(str),  # Color by cluster label (convert to string for categorical coloring)
    text='OT',  # Text labels for Operating Time
    title='DBSCAN Clustering of Operating Current (OC) vs Peak Current (PC)',
    labels={'cluster': 'Cluster'}  # Label the color legend
)

# Update trace settings for text position
fig.update_traces(textposition="top center")

# Show the plot
fig.show()
```
